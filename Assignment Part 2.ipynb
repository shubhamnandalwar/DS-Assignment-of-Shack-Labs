# Data Gathering

```python
amzon=pd.read_csv("amz_com-ecommerce_sample.csv",encoding= 'unicode_escape')
amz=pd.DataFrame(amz)
amz_df = amz.rename(columns={"product_name":"Product name in Amazon","retail_price":"Retail Price in Amazon","discounted_price":'Discounted Price in Amazon'}, index={'ONE': 'Row_1'})
amz_df.head()
```

```python
flipkart=pd.read_csv(r"flipkart_com-ecommerce_sample.csv")
flp=pd.DataFrame(flipkart)
flp_df = flp.rename(columns={"product_name":"Product name in Flipkart","retail_price":"Retail Price in Flipkart","discounted_price":'Discounted Price in Flipkart'}, index={'ONE': 'Row_1'})
flp_df.head()
```

```python
df=pd.concat([amz_df,flp_df],axis = 1, join = "inner").head()
df.shape
```

```python
df.columns
```

```python
# Converting Dataframe into Excel
```

```python
match_product=df.loc[:,['Product name in Flipkart','Retail Price in Flipkart','Discounted Price in Flipkart','Product name in Amazon','Retail Price in Amazon','Discounted Price in Amazon']] 
match_product.to_csv("similar product of Flp and Amz.csv")
```
